```python
import numpy as np
import matplotlib.pyplot as plt
```

# Example demonstrating Monte Carlo technique
- Estimating value of pi via Monte Carlo
- Value of pi calculated by finding the ratio of points that satisfy equation of a circle(fall within the circle) and outside in a unit square. The unit square is just one quarter of the circle

```python
np.random.seed(0)
N = 2000
r = 1

# Generate random x,y coordinates inside the square with side=r.
x = np.random.uniform(0, r, N)
y = np.random.uniform(0, r, N)

# Distance/radius list made by each co-ordinate to center/origin
rAll = np.sqrt(x ** 2 + y ** 2)

# To color points inside/outside the circle
pointSizesForMasking = 10 * np.ones(N)

pointsInsideCircle  = np.ma.masked_where(rAll < r,  pointSizesForMasking)
pointsOutsideCircle = np.ma.masked_where(rAll >= r, pointSizesForMasking)

# Plot showing points inside/outside circle
axes=plt.gca()
axes.set_aspect('equal')
plt.scatter(x, y, s=pointsInsideCircle, marker='o')
plt.scatter(x, y, s=pointsOutsideCircle, marker='o')
xr = np.linspace(0, 1, 100)
yr = np.sqrt(1 - xr**2)
plt.plot(xr, yr, c='k')
plt.title("Randomly generated points in a unit square")
```

```python
ctr = 0
for i in range(N):
    # Check if point is inside the circle
    if np.sqrt(x[i] ** 2 + y[i] ** 2) <= 1:
        ctr = ctr + 1

# 4 * ratio of orange dots as this is only quarter of a circle
calculatedPi = str(4 * ctr / N)

print("pi value estimated using Mone Carlo technique: " + calculatedPi)

```

```python
np.random.seed(100)
N = 100
Y = []
X = range(0, N)

Y.append(np.random.normal())
for t in range(1, N):
    Y.append(Y[t-1] + np.random.normal())
    

plt.plot(X, Y)
plt.title('Random walk - A simple Markov chain')
```
